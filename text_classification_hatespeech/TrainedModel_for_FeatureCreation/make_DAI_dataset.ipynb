```python
import pandas as pd

from transformers import AutoTokenizer
```

```python
df_train = pd.read_csv('tmp/hatespeech_train_zeroshots.csv')
df_test = pd.read_csv('tmp/hatespeech_test_zeroshots.csv')
df_train.shape, df_test.shape
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
## label別、各指標の平均値
df_train[['label','negative1','negative2','malice','hate','discrimination']].groupby('label').mean()
```

```python
## label別、各指標の平均値
df_test[['label','negative1','negative2','malice','hate','discrimination']].groupby('label').mean()
```

```python
tokenizer = AutoTokenizer.from_pretrained('tohoku-nlp/bert-base-japanese-whole-word-masking')
tokenizer
```

```python
df_train['text_token_tohokuBertBase'] = df_train['text'].apply(lambda x: " ".join([tkn for tkn in tokenizer.tokenize(x)]))
df_test['text_token_tohokuBertBase'] = df_test['text'].apply(lambda x: " ".join([tkn for tkn in tokenizer.tokenize(x)]))
```

```python
df_train.head()
```

```python
df_train.to_csv('tmp/hatespeech_train_zeroshots_tkn.csv', index=False)
df_test.to_csv('tmp/hatespeech_test_zeroshots_tkn.csv', index=False)
```
