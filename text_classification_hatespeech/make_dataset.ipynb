```python
df =pd.read_csv('tmp/data/train.csv')
df
```

```python
df['label'].value_counts()
```

```python
df_train, df_test = train_test_split(df, test_size=600, stratify=df['label'], random_state=123)

df.shape, df_train.shape, df_test.shape, df_test.shape[0]/ df_train.shape[0]
```

```python
df_train['label'].value_counts()
```

```python
df_test['label'].value_counts()
```

```python
np.random.seed(seed=12345)
hold = np.random.choice([0,1,2,3,4], size=df_train.shape[0])
df_train['hold'] = hold

print(df_train['hold'].value_counts(normalize=True))

pd.crosstab(df_train['hold'], df_train['label'])#, normalize='index')
```
