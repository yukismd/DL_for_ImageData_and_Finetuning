```python
dirs_test = ["preds_hatespeech-round3_grad_accumulation_1.fold0.predict_fold0",
     "preds_hatespeech-round3_grad_accumulation_1.1.fold0.predict_fold0",
     "preds_hatespeech-round3_grad_accumulation_1.2.fold0.predict_fold0",
     "preds_hatespeech-round3_grad_accumulation_1.3.fold0.predict_fold0",
     "preds_hatespeech-round3_grad_accumulation_1.4.fold0.predict_fold0"]
```

```python
names = ["hatespeech-round3_grad_accumulation_1",
     "hatespeech-round3_grad_accumulation_1.1.",
     "hatespeech-round3_grad_accumulation_1.2",
     "hatespeech-round3_grad_accumulation_1.3",
     "hatespeech-round3_grad_accumulation_1.4"]
```

```python
df_res_train = pd.DataFrame()
for d,n in zip(dirs_train, names):
    f = 'tmp/train_res/'+d+'/test_predictions.csv'
    tmp = pd.read_csv(f)
    print(f, ' --- ', tmp.shape)
    tmp = tmp.rename(columns={'pred_label':n})
    tmp = tmp.drop(columns=['id','source','text','label','hold'])
    df_res_train = pd.concat([df_res_train, tmp], axis=1)
    print(df_res_train.shape)
```

```python
df_res_train
```

```python
df_res_test = pd.DataFrame()
for d,n in zip(dirs_test, names):
    f = 'tmp/test_res/'+d+'/test_predictions.csv'
    tmp = pd.read_csv(f)
    print(f, ' --- ', tmp.shape)
    tmp = tmp.rename(columns={'pred_label':n})
    tmp = tmp.drop(columns=['id','source','text','label'])
    df_res_test = pd.concat([df_res_test, tmp], axis=1)
    print(df_res_test.shape)
```

```python
df_res_test
```

```python
df_train = pd.read_csv('../tmp/hatespeech_train.csv')
df_train
```

```python
df_train = pd.concat([df_train, df_res_train], axis=1)
df_train.shape
```

```python
df_test = pd.concat([df_test, df_res_test], axis=1)
df_test.shape
```
