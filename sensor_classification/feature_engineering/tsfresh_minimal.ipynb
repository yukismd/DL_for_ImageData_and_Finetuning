```python
settings = MinimalFCParameters()
settings
```

```python
df_basic_features = pd.DataFrame()

for k,v in dic_features.items():
    print('>> add {} which shape is {}'.format(k, v.shape))
    tmp = extract_features(v, column_id="id", column_sort="time", default_fc_parameters = settings)
    df_basic_features = pd.concat([df_basic_features, tmp], axis=1)
    print('current shape is {}'.format(df_basic_features.shape))

df_basic_features.shape
```
