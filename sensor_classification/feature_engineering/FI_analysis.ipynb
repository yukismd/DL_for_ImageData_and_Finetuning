```python
#df = pd.read_csv('tmp/SenCls_minimal_features_default/data---original-feature-importance.csv', skiprows=2)
df = pd.read_csv('tmp/SenCls_minimal_features_default/data---original-feature-importance__No5.csv', skiprows=2)
df
```

```python
df['features'] = [i.split('__')[0] for i in df['label']]
df['stats'] = [i.split('__')[1] for i in df['label']]
```

```python
#df.pivot()
```

```python
df_f = df.groupby('features')['value'].sum().reset_index().sort_values('value', ascending=False)
df_f
```

```python
df_s = df.groupby('stats')['value'].sum().reset_index().sort_values('value', ascending=False)
df_s
```

```python
#df_f.to_csv('tmp/SenCls_minimal_features_default/fi_features.csv', index=False)
#df_s.to_csv('tmp/SenCls_minimal_features_default/fi_stats.csv', index=False)

df_f.to_csv('tmp/SenCls_minimal_features_default/fi_features_No5.csv', index=False)
df_s.to_csv('tmp/SenCls_minimal_features_default/fi_stats_No5.csv', index=False)
```
