```python
# Create the scatter plot using seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Model Type", y="Validation mAP", hue="Backbone", palette="Set1", s=100)

# Improve plot aesthetics
plt.title("Validation IoU vs Architecture, colored by Backbone")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
