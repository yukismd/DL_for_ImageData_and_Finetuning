```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#from PIL import Image

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tmp/DAI_document_classification_TEST/data.csv')
df.head()
```

```python
df_resume = df[df['class']=='resume']
df_email = df[df['class']=='email']
df_sp = df[df['class']=='scientific_publication']
```

```python
img_path = 'tmp/DAI_document_classification_TEST/'

df_resume_path = [img_path+i for i in df_resume['image']]
df_email_path = [img_path+i for i in df_email['image']]
df_sp_path = [img_path+i for i in df_sp['image']]
```

```python
image_paths = df_resume_path

fig, axes = plt.subplots(1, len(image_paths), figsize=(15, 5))
for ax, image_path in zip(axes, image_paths):
    img = mpimg.imread(image_path)
    ax.imshow(img)
    ax.axis('off')  # 

plt.gray()
plt.tight_layout()
plt.savefig("display_images/resume.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
image_paths = df_email_path

fig, axes = plt.subplots(1, len(image_paths), figsize=(15, 5))
for ax, image_path in zip(axes, image_paths):
    img = mpimg.imread(image_path)
    ax.imshow(img)
    ax.axis('off')  # 

plt.gray()
plt.tight_layout()
plt.savefig("display_images/email.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
image_paths = df_sp_path

fig, axes = plt.subplots(1, len(image_paths), figsize=(15, 5))
for ax, image_path in zip(axes, image_paths):
    img = mpimg.imread(image_path)
    ax.imshow(img)
    ax.axis('off')  # 

plt.gray()
plt.tight_layout()
plt.savefig("display_images/scientific_publication.png", dpi=300, bbox_inches='tight')
plt.show()
```
