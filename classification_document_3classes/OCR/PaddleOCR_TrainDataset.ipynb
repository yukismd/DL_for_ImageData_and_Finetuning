```python
import pandas as pd
#from PIL import Image
from paddleocr import PaddleOCR

import pickle
```

```python
!ls ../tmp/DAI_document_classification_TRAIN
```

```python
df = pd.read_csv('../tmp/DAI_document_classification_TRAIN/data.csv')
print(df.shape)
df.head()
```

```python
img_dir = '../tmp/DAI_document_classification_TRAIN/'
```

```python
ocr = PaddleOCR(use_angle_cls=True, lang='en')
```

```python
%%time

all_scripts = {}
for img in df['image']:
    img_name = img.split('/')[1]
    img_path = img_dir + img
    print('START >> ', img_path)
    
    result = ocr.ocr(img_path, cls=True)   # Detection + Classification + Recognition
    
    script = []
    
    for idx in range(len(result)):
        res = result[idx]
        for line in res:
            #print(line[1][0])
            script.append(line[1][0])   # 
            
    all_scripts[img_name] = script

len(all_scripts)
```

```python
path = 'outputs/ocr_res_train.pkl'
with open(path, mode='wb') as f:
    pickle.dump(all_scripts,f)
```
