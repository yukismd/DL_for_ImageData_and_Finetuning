```python
image_dir = '../tmp/DAI_document_classification_TEST/images'

images = ['doc_000016.png', 'doc_000101.png', 'doc_000297.png', 'doc_000513.png', 'doc_000655.png',
          'doc_000045.png', 'doc_000130.png', 'doc_000349.png', 'doc_000550.png', 'doc_000674.png',
          'doc_000070.png', 'doc_000175.png', 'doc_000363.png', 'doc_000551.png', 'doc_000745.png',
          'doc_000076.png', 'doc_000248.png', 'doc_000402.png', 'doc_000558.png', 'doc_000802.png',
          'doc_000089.png', 'doc_000279.png', 'doc_000485.png', 'doc_000584.png', 'doc_000832.png',
          'doc_000097.png', 'doc_000281.png', 'doc_000509.png', 'doc_000596.png', 'doc_000864.png']
```

```python
load_dotenv('.env')

client = OpenAI(api_key=os.getenv('openai-key-20241027'))
client
```

```python
%%time

responses = {}

vision_model = "gpt-4o-mini"
prompt = "Extract the type of the image, categorizing it as 'email', 'resume' or 'scientific_publication'. Type:"

# Function to encode the image
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")

for img in images:
    print('>>>>> ', img)
    
    # Path to your image
    image_path = os.path.join(image_dir, img)
    
    # Getting the base64 string
    base64_image = encode_image(image_path)
    #print(type(base64_image), len(base64_image))

    response = client.chat.completions.create(
        model=vision_model,
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": prompt,
                    },
                    {
                        "type": "image_url",
                        "image_url": {"url": f"data:image/jpeg;base64,{base64_image}"},
                    },
                ],
            }
        ],
    )
    responses[img] = response.choices[0].message.content
    print(responses[img])
```

```python
responses
```

```python
df = pd.DataFrame(list(responses.items()), columns=['image_id', prompt])
df
```

```python
df.to_csv('prediction/gpt-4o-mini_result.csv', index=False)
```
