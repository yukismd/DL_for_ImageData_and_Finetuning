```python
df_train=pd.read_csv('tmp/HT/cable_classify_15classes_HT/train.csv')
df_train
```

```python
df_train['label'].value_counts().sort_values(ascending=False)
```

```python
df_train['label'].value_counts().sort_values(ascending=False).index
```

```python
val = df_train['label'].value_counts().sort_values(ascending=False).index

mapping = {val[i]:i for i in range(len(val))}
mapping
```

```python
df_train['label_idx'] = df_train['label'].map(mapping)
df_train.head()
```

```python
list_0_6 = [i for i in range(7)]
np.random.seed(seed=123)
df_train['fold'] = np.random.choice(list_0_6, size=df_train.shape[0])
df_train.head()
```

```python
df_train = df_train.rename(columns={'label':'label_original', 'label_idx':'label'})
df_train
```

```python
df_test=pd.read_csv('tmp/HT/cable_classify_15classes_HT/test.csv')
df_test
```

```python
df_test['label_idx'] = df_test['label'].map(mapping)
df_test = df_test.rename(columns={'label':'label_original', 'label_idx':'label'})
df_test
```

```python
df_train.to_csv('tmp/PT/cable_classify_15classes_PT/train.csv', index=False)
df_test.to_csv('tmp/PT/cable_classify_15classes_PT/test.csv', index=False)
```
