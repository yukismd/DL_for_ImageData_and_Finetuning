```python
import torch
from transformers import AutoModel, AutoTokenizer

import pandas as pd
```

```python
# Set up the model and tokenizer
model_path = 'h2oai/h2ovl-mississippi-2b'
model = AutoModel.from_pretrained(
    model_path,
    torch_dtype=torch.bfloat16,
    low_cpu_mem_usage=True,
    trust_remote_code=True).eval().cuda()
tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True, use_fast=False)
generation_config = dict(max_new_tokens=1024, do_sample=True)
```

```python
!ls images
```

```python
path = 'images/'

images =['driveRecord1.jpg','driveRecord2.jpg','driveRecord3.jpg','driveRecord4.jpg','driveRecord5.jpg',
         'driveRecord6.jpg','driveRecord7.jpg','driveRecord8.jpg','driveRecord9.png','driveRecord10.jpg']
```

```python
%%time

#question = '<image>\nDoes crosswalk appear in the image?'
question = '<image>\nDoes pedestrian crossing appear in the image?'

answer = []
for img in images:
  image_file = path+img
  response, history = model.chat(tokenizer, image_file, question, generation_config, history=None, return_history=True)
  print(img , ' : ', response)
  answer.append(response)

question_1 = question
answer_1 = answer
print('----- DONE -----')
```
