```python
df = pd.read_csv('tmp/train_v2.csv')
print(df.shape)
df.head()
```

```python
df['target'].value_counts()
```

```python
df['skf_fold'].value_counts()
```

```python
pd.crosstab(df['skf_fold'], df['target'])
```

```python
# 100ずつ取得 -> 合計5000

# 240 -> 12000
```

```python
df['stratify'] = [f'{i}-{j}' for i,j in zip(df['skf_fold'],df['target'])]
```

```python
df['stratify'].value_counts()
```

```python
gdf = df.groupby('stratify')
```

```python
df_sample = gdf.apply(lambda x: x.sample(n=240))
df_sample.shape
```

```python
df_sample
```

```python
df_sample = df_sample.reset_index(drop=True)
df_sample
```

```python
#df_sample = df_sample.sort_values('id')
```

```python
# 10000 -> train
# 2000 -> test
```

```python
gdf2 = df_sample.groupby('stratify')
df_test = gdf2.apply(lambda x: x.sample(n=40))
df_test.shape
```

```python
df_sample['test_flg'] = df_sample['id'].isin(df_test['id'])*1
```

```python
df_sample
```

```python
df_sample['test_flg'].value_counts()
```

```python
df_sample = df_sample.sort_values('id')
df_sample
```

```python
pd.crosstab(df_sample['skf_fold'], df_sample['target'])
```

```python
df_dai = df_sample.copy()
df_ht = df_sample.copy()
```

## DAI

```python
df_dai['images'] = [f'images/{i}.png' for i in df_dai['id']]
```

```python
df_dai
```

```python
df_dai_train = df_dai[df_dai['test_flg']==0]
df_dai_test = df_dai[df_dai['test_flg']==1]
df_dai_train.shape, df_dai_test.shape
```

```python
df_dai_train = df_dai_train[['id', 'images', 'target', 'label']]
df_dai_test = df_dai_test[['id', 'images', 'target', 'label']]
df_dai_train.shape, df_dai_test.shape
```

```python
df_dai_train = df_dai_train.rename(columns={'label':'target_label'})
df_dai_test = df_dai_test.rename(columns={'label':'target_label'})
```

```python
df_dai_train
```

```python
# from tmp/images
# to tmp/cifar10_sample_DAI_train/images   &   tmp/cifar10_sample_DAI_test/images
```

```python
for im in df_dai_train['images']:
    from_dir = 'tmp/'
    to_dir = 'tmp/cifar10_sample_DAI_train/'
    shutil.copy(from_dir+im, to_dir+im)
```
