```python
import pandas as pd
from sklearn.metrics import root_mean_squared_error
```

```python
!ls ../data
```

```python
df = pd.read_csv('../data/walmart_ts_6_fcst_grp_test.csv')
print(df.shape)
df.head()
```

```python
!ls ../Chronos_results
```

```python
df_c_0 = pd.read_csv('../Chronos_results/ChronosBolt_0shot.csv')
df_c_ft = pd.read_csv('../Chronos_results/ChronosBolt_FT.csv')
df_c_0_cv = pd.read_csv('../Chronos_results/ChronosBolt_0Shot_Covariate.csv')
df_c_ft_cv = pd.read_csv('../Chronos_results/ChronosBolt_FT_Covariate.csv')
print(df_c_0.shape, df_c_ft.shape, df_c_0_cv.shape, df_c_ft_cv.shape)
df_c_0.head()
```

```python
df['mean_Chronos-0shot'] = df_c_0['mean']
df['mean_Chronos-FT'] = df_c_ft['mean']
df['mean_Chronos-0shot-covariate'] = df_c_0_cv['mean']
df['mean_Chronos-FT-covariate'] = df_c_ft_cv['mean']
df.head()
```

```python
# 全体

# 0 Shot
print(root_mean_squared_error(y_true=df['Weekly_Sales'], y_pred=df['mean_Chronos-0shot']))

# FT
print(root_mean_squared_error(y_true=df['Weekly_Sales'], y_pred=df['mean_Chronos-FT']))

# 0 Shot, covariate
print(root_mean_squared_error(y_true=df['Weekly_Sales'], y_pred=df['mean_Chronos-0shot-covariate']))

# FT, covariate
print(root_mean_squared_error(y_true=df['Weekly_Sales'], y_pred=df['mean_Chronos-FT-covariate']))
```

```python
# グループ別

# 0 shot
rmse_by_category = df.groupby(['Store','Dept']).apply(lambda g: root_mean_squared_error(g['Weekly_Sales'], g['mean_Chronos-0shot']))
display(rmse_by_category.reset_index(name="RMSE"))

# FT
rmse_by_category = df.groupby(['Store','Dept']).apply(lambda g: root_mean_squared_error(g['Weekly_Sales'], g['mean_Chronos-FT']))
display(rmse_by_category.reset_index(name="RMSE"))

# 0 shot, covariate
rmse_by_category = df.groupby(['Store','Dept']).apply(lambda g: root_mean_squared_error(g['Weekly_Sales'], g['mean_Chronos-0shot-covariate']))
display(rmse_by_category.reset_index(name="RMSE"))

# FT, covariate
rmse_by_category = df.groupby(['Store','Dept']).apply(lambda g: root_mean_squared_error(g['Weekly_Sales'], g['mean_Chronos-FT-covariate']))
display(rmse_by_category.reset_index(name="RMSE"))
```
